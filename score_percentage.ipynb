# **Prediction using Supervised ML**
*Predict the percentage of a student based on the number of study hours.
This is a simple linear regression task involving two variables.*

Import the libraries that are required for implement the prediction.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Load the data from the url provided http://bit.ly/w-data 

Read the data using read_csv

```python
url = ("http://bit.ly/w-data")
dataload = pd.read_csv(url)
print('Data imported!')
```

```python
dataload.head()
```

Plot the data for visualisation.

```python
dataload.plot(x='Hours', y='Scores', style='o')
plt.title('Hours Vs Percentage')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
```

View statictical data

```python
dataload.describe()
```

Divide the data into attributes and labels

```python
X = dataload.iloc[:, :-1].values
y = dataload.iloc[:, 1].values
```

Use SciKit Learn's train_test_split() for training the data and test sets.

*Split the data using train_test_split()*

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

*Train the algorithm using Linear Regression*

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
print("Training Completed!")
```

Plot the data using the previously trained data

```python
line = reg.coef_*X+reg.intercept_
plt.scatter(X, y)
plt.plot(X, line, color='green')
plt.show()
```

Compare the actual model with the predicted model

```python
pred = reg.predict(X_test)
dif = pd.DataFrame({"Actual": y_test, "Predicted": pred})
dif
```

Visualizing the difference between actual data and predicted data

```python
plt.scatter(x=X_test, y=y_test, color='red')
plt.plot(X_test, pred, color='blue')
plt.title("Actual Vs Predicted", size=22)
plt.xlabel("Hours Studied")
plt.ylabel("Percentage Score", size=12)
plt.show()
```

Test the model with 9.25 hours.

```python
hrs = [9.25]
score = reg.predict([hrs])
print('Score: ', score[0])
```

### *Therefore the model shows if a student studies for 9.25 hrs/day he/she will be scecuring 93.69173248737539 percentage score.*
